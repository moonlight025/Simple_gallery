```python
import requests
from bs4 import BeautifulSoup
from requests.models import LocationParseError
import time
```

```python
first_html = requests.get('https://www.ss.net.tw/')
soup = BeautifulSoup(first_html.text,'html.parser')
all_div = soup.find_all('div', class_ = 'col mb-3')

target_div = all_div[21:]
print(soup.prettify())
```

```python
painter_link = []
amount = []
for t in target_div:
    tmp = t.find('a')
    painter_link.append(tmp.get('href'))
    tmp = t.find('small', class_ = 'ml-md-1')
    amount.append(tmp.text[1:-1])
# print(data)
# print(painter_link)
```

```python
picture_link = []
for i in range(len(amount)):
    for j in range(int(amount[i])//24 + 1):
        tmp_link = str(painter_link[i] + f"~{j+1}")
        picture_html = requests.get(tmp_link)
        if picture_html.status_code!=200:
            print(picture_html.status_code)
        picture_soup = BeautifulSoup(picture_html.text,'html.parser')
        div = picture_soup.find_all('div', class_ = 'row row-cols-2 row-cols-sm-3 row-cols-md-4 row-cols-lg-5 row-cols-xl-6 align-items-end')
        tmp = div[0].find_all('div',class_ = 'col')
        for k in range(len(tmp)):
            link = tmp[k].find('a')
            picture_link.append(link.get('href'))
            time.sleep(0.1)
print(picture_link)
```
